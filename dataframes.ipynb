```python
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)

print(df)


```

```python
# Copy the code from the lesson to create a dataframe full of student grades.

print('1a')
# Create a column named passing_english that indicates whether each student has a passing grade in english.

# 
df['passing_english'] = df['english'] >= 70

print(df)
```

```python
print('1b')
# Sort the english grades by the passing_english column. How are duplicates handled/

df.sort_values('passing_english')

print('duplicated are handled ascending')
```

```python
print('1c')
# Sort the english grades first by passing_english and then by student name. All the students that are failing 
# english should be first, and within the students that are failing english they should be ordered alphabetically. 
# The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)

df.sort_values(['passing_english','name'])
```

```python
print('1d')
# Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in 
# the last step.

df.sort_values(['passing_english','english'])
```

```python
print('1e')
# Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average 
# of the math, english, and reading grades.


# df['overall_grade'] = df[['math', 'english', 'reading']].mean(axis=1).round(2)
# .mean() produces mean of columns have to include the axist to mean across columns

# another way
df['overall_grad'] = round((df.math + df.english + df.reading)/3,1)


# Display the DataFrame with the new overall_grade column
print(df)
```

```python
Mammals = data('Mammals')

# How many rows and columns are there?

print(Mammals.shape)
# What are the data types?

print(Mammals.dtypes)
# Summarize the dataframe with .info and .describe
print('Mammals info')
print(Mammals.info())


print('Mammals description')
print(Mammals.describe())

# View the column names
column_names = Mammals.columns
print("Column names:")
print(column_names)



What is the the weight of the fastest animal?

# .max find fastest
max_speed = Mammals['speed'].max()

# bool mask 
fastest_animal_mask = Mammals['speed'] == max_speed

# Use the mask to select 
fastest_animal_data = Mammals[fastest_animal_mask]

# use fast data to seach for weight
fastest_animal_weight = fastest_animal_data['weight'].values[0]

# Print the weight of the fastest animal
print(f"The weight of the fastest animal is {fastest_animal_weight} kg")

# another way
# Find the maximum speed in the dataset
max_speed = Mammals['speed'].max()

# Use .loc to locate the weight of the fastest animal
fastest_animal_weight = Mammals.loc[Mammals['speed'] == max_speed, 'weight'].values[0]

# Print the weight of the fastest animal
print(f"The weight of the fastest animal is {fastest_animal_weight} kg")

# another way

fastest_animal_weight = Mammals[Mammals['speed'] == Mammals['speed'].max()]['weight'].values[0]
print(f"The weight of the fastest animal is {fastest_animal_weight} kg")


# What is the overal percentage of specials?
# (total specials/all mammals)*100
percentage_specials = (Mammals['specials'].sum() / len(Mammals)) * 100

# Print the result
print(f"The overall percentage of specials is {percentage_specials:.2f}%")


# How many animals are hoppers that are above the median speed? What percentage is this?
# median speed across all mammals, filter by hoppers then by > median speed, (result/total >median speed)*100

median_speed = Mammals['speed'].median()

print(f'median speed is: {median_speed}')


hoppers = Mammals[Mammals['hoppers'] == True]

print(f'number of hoppers: {hoppers}')
hopper_count = len(hoppers)  # Count of hoppers
print(f'# of hoppers are: {hopper_count}')

                  
hoppers_above_median = hoppers[hoppers['speed'] > median_speed]

# Count above med
num_hoppers_above_median = len(hoppers_above_median)

# percentage
percentage_hoppers_above_median = (num_hoppers_above_median / len(Mammals)) * 100

# Print the results
print(f"Number of hoppers above the median speed: {num_hoppers_above_median}")
print(f"Percentage of hoppers above the median speed: {percentage_hoppers_above_median:.2f}%")


```
