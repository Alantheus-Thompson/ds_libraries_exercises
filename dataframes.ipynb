```python
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)

print(df)


```

```python
# Copy the code from the lesson to create a dataframe full of student grades.

print('1a')
# Create a column named passing_english that indicates whether each student has a passing grade in english.

# 
df['passing_english'] = df['english'] >= 70

print(df)




```

```python
print('1b')
# Sort the english grades by the passing_english column. How are duplicates handled/

df.sort_values(by='passing_english')

print('duplicated are handled ascending')
```

```python
print('1c')
# Sort the english grades first by passing_english and then by student name. All the students that are failing 
# english should be first, and within the students that are failing english they should be ordered alphabetically. 
# The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)

df.sort_values(by=['passing_english','name'])
```

```python
print('1d')
# Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in 
# the last step.

df.sort_values(by=['passing_english','english'])
```

```python
print('1e')
# Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average 
# of the math, english, and reading grades.


df['overall_grade'] = df[['math', 'english', 'reading']].mean(axis=1).round(2)

# Display the DataFrame with the new overall_grade column
print(df)
```

```python
print('2')
# Load the mpg dataset. Read the documentation for the dataset and use it for the following questions:

from pydataset import data
# data('mpg', show_doc=True) # view the documentation for the dataset
mpg = data('mpg') # load the dataset and store it in a variable

# How many rows and columns are there?

print(mpg.shape)
num_rows, num_columns = mpg.shape
print(f"Number of rows: {num_rows}, Number of columns: {num_columns}")

# What are the data types of each column?
print("Data types of each column:")
print(mpg.dtypes)

# Summarize the dataframe with .info and .describe

print('dataframe info')
print(mpg.info())

print('dataframe description')
print(mpg.describe())

# Rename the cty column to city and hwy to highway

print('name edits cty to city hwy to highway')
mpg.rename(columns={'cty': 'city', 'hwy': 'highway'}, inplace=True)
print([mpg])

# Do any cars have better city mileage than highway mileage?

print('better city mileage than highway')
better_city_mileage = mpg[mpg['city'] > mpg['highway']]
if better_city_mileage.empty:
    print("No cars have better city mileage than highway mileage.")
else:
    print("Cars with better city mileage than highway mileage:")
    print(better_city_mileage)

# Create a column named mileage_difference this column should contain the difference between highway and city 
# mileage for each car.

print('mileage difference between highway and city')
mpg['mileage_difference'] = mpg['highway'] - mpg['city']
print([mpg['mileage_difference']])

# Which car (or cars) has the highest mileage difference?

max_mileage_difference = mpg['mileage_difference'].max()
cars_with_highest_difference = mpg[mpg['mileage_difference'] == max_mileage_difference]
print("Car(s) with the highest mileage difference:")
print(cars_with_highest_difference)

# Which compact class car has the lowest highway mileage? The best?
#variable to find compact cars
compact_cars = mpg[mpg['class'] == 'compact']
# variable to find lowest highway in compact variable using .min
lowest_highway_mileage = compact_cars['highway'].min()

# variable to find highest in compact using .max
best_highway_mileage = compact_cars['highway'].max()

## NOTE curly brackets within f print to insert variables.  do not indicate dictionaries
print(f'The compact car with the lowest highway mileage is: {lowest_highway_mileage} mpg.')
print(f'The compact car with the best highway mileage has {best_highway_mileage} mpg.')


# Create a column named average_mileage that is the mean of the city and highway mileage.
mpg['average_mileage'] = mpg[['city','highway']].mean(axis=1)
print(mpg)

# Which dodge car has the best average mileage? The worst?
# variable looking across df column manufacturer the same as dodge wraped in df output
dodge_cars = mpg[mpg['manufacturer'] == 'dodge']

# variable best avg looking across dodge cars avg mileage variable looking for max
best_avg_mileage = dodge_cars['average_mileage'].idxmax()

# variable worst avg looking across dodge cars avg vari looking for min
worst_avg_mileage = dodge_cars['average_mileage'].idxmin()

# vari best dodge car based on dodge cars using .loc which allows filtering of sorts on rows 
best_dodge_car = dodge_cars.loc[best_avg_mileage]
worst_dodge_car = dodge_cars.loc[worst_avg_mileage]
print(f"The Dodge car with the best average mileage is:")
print(best_dodge_car)
print(f"The Dodge car with the worst average mileage is:")
print(worst_dodge_car)
```

```python
Mammals = data('Mammals')

# How many rows and columns are there?

print(Mammals.shape)
# What are the data types?

print(Mammals.dtypes)
# Summarize the dataframe with .info and .describe
print('Mammals info')
print(Mammals.info())


print('Mammals description')
print(Mammals.describe())

# View the column names
column_names = Mammals.columns
print("Column names:")
print(column_names)



What is the the weight of the fastest animal?

# .max find fastest
max_speed = Mammals['speed'].max()

# bool mask 
fastest_animal_mask = Mammals['speed'] == max_speed

# Use the mask to select 
fastest_animal_data = Mammals[fastest_animal_mask]

# use fast data to seach for weight
fastest_animal_weight = fastest_animal_data['weight'].values[0]

# Print the weight of the fastest animal
print(f"The weight of the fastest animal is {fastest_animal_weight} kg")

# another way
# Find the maximum speed in the dataset
max_speed = Mammals['speed'].max()

# Use .loc to locate the weight of the fastest animal
fastest_animal_weight = Mammals.loc[Mammals['speed'] == max_speed, 'weight'].values[0]

# Print the weight of the fastest animal
print(f"The weight of the fastest animal is {fastest_animal_weight} kg")

# another way

fastest_animal_weight = Mammals[Mammals['speed'] == Mammals['speed'].max()]['weight'].values[0]
print(f"The weight of the fastest animal is {fastest_animal_weight} kg")


# What is the overal percentage of specials?
# (total specials/all mammals)*100
percentage_specials = (mammals['specials'].sum() / len(mammals)) * 100

# Print the result
print(f"The overall percentage of specials is {percentage_specials:.2f}%")


# How many animals are hoppers that are above the median speed? What percentage is this?
# median speed across all mammals, filter by hoppers then by > median speed, (result/total >median speed)*100

median_speed = Mammals['speed'].median()


hoppers = Mammals[Mammals['hoppers'] == 1]
                  
hoppers_above_median = [hoppers['speed'] > median_speed]

# Count the number of animals that meet the criteria
num_hoppers_above_median = len(hoppers_above_median)

# Calculate the percentage
percentage_hoppers_above_median = (num_hoppers_above_median / len(Mammals)) * 100

# Print the results
print(f"Number of hoppers above the median speed: {num_hoppers_above_median}")
print(f"Percentage of hoppers above the median speed: {percentage_hoppers_above_median:.2f}%")


```
