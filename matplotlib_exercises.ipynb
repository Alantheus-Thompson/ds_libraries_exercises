```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-20,20)
```

```python
# Create and label 4 separate charts with the following equations (choose a range for x that makes sense):
# y = sqrt x
# y = x**3
# y = 2**x
# y = 1/(x+1)

# You can use functions from the math module to help implement some of the equations above.
```

```python
y = np.sqrt(x)

plt.plot(x,y)
plt.title('Sqrt of X')
plt.show()
```

```python
y = x**3
plt.plot(x,y)
plt.title('X Cubed')
plt.show()
```

```python
y = 2**x
plt.plot(x,y)
plt.title('2 to the Power of X')
plt.show()
```

```python
y = 1/(x+1)
plt.plot(x, y)
plt.title('1 over X+1')
plt.show()
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
```

```python
# Note to self.  Received error message on x operators need list comprehension ###
# x = [0,1,2,3,4,5]

# y1 = [xi**0.5 for xi in x]
# y2 = [xi**3 for xi in x]
# y3 = [2**xi for xi in x]
# y4 = [1 / (xi + 1) for xi in x]

# with numpy however ##
x = np.linspace(-20, 20) 

y1 = np.sqrt(x)
y2 = x**3
y3 = 2**x
y4 = 1 / (x + 1)

plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title('Sqrt X')

plt.subplot(2,2,2)
plt.plot(x, y2)
plt.title('X Cubed')

plt.subplot(2,2,3)
plt.plot(x, y3)
plt.title('2 to the X power')

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title('1 divided by X+1')

plt.suptitle('Subplots Demo')

plt.show()


```
